```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
%matplotlib inline 
```

```python
shamiri_data = pd.read_csv('shamiri_imputed_dataset.csv', index_col='ParticipantID')
shamiri_data
```

<p>From the initial description of the data, we draw the following observations:</p>
<ul>
    <li>The data is from 658 participants of different ages from different schools. We have 80, 87, 122, 157, and 212 participants from Starays, Elite, AHS, AGHS, and Olympic schools respectively. </li>
    <li>Majority of the respondents were from the "majority" tribe.</li>
    <li>There were 338 female respondents against 320 male respondents participating in the study.</li>
    <li>Most of the respondents were between the ages of 14-18 years.</li>
    <li>279, 212, 167 respondents were from rich, medium, and poor school environments respectively.</li>
    <li>The data has no null values.</li>
    <li>All the columns are in their correct data types.</li>
    <li>The maximum value in column 'MSSS10' is 8. Further investigation notes that only one person allegedly scored this value. This appears to be a mistake. We will need to expunge that whole row, to retain the integrity of the data. We have assumed that the MSSS scale is 0-7 as the others.</li>
</ul> 
<p>From these observations we conclude that the data is ready for further analysis.</p>

```python
#We find the row where column 'MSSS10' value is 8 and remove it

p = shamiri_data.index[shamiri_data['MSSS10'] == 8].tolist()  #We find the row
shamiri_data.drop(p, inplace = True) #We delete the row
shamiri_data.describe()
```

<h1>Analysis of Patient Health Questionnaire(PHQ)</h1>
<p>To analyze PHQ data we will need to create a new and exclusive DataFrame of only PHQ data. We will also need to create a new column for the individual scores.</p>

```python
PHQ_data = shamiri_data.iloc[ :, :8] #We slice the original dataframe to remain only with PHQ data
score = PHQ_data.sum(axis=1)  #We then find the individual scores of each participant
PHQ_data['PHQ_score'] = score  #We then create a new column showing the score of each participant
PHQ_data
```

```python
print(PHQ_data['PHQ_score'].min(), PHQ_data['PHQ_score'].max())    #We confirm whether the scores in the score column are within range
```

<p>We then create bins to classify the individual scores into their respective depressive symptoms diagnosis. We know that the cutoffs for mild, moderate, moderately-severe, and severe depression are 5, 10, 15, and 20 respectively. We also know that PHQ scale runs from 0 to 24.</p>

```python
bins = np.arange(0, 26, 5)
scale = ['mild', 'moderate', 'moderately-severe', 'severe', 'more-severe']
PHQ_data['PHQ_range'] = pd.cut(PHQ_data['PHQ_score'], bins=bins, labels=scale, include_lowest=True)
PHQ_data
```

```python
PHQ_data['PHQ_range'].value_counts(normalize=True, sort=False)   #Proportion of the total
```

<p>The above output shows that:</p> 
<ul>
    <li>Moderate depression cases was leading with 34% of the total cases while the more severe cases of depression was trailing with a mere 2% of the total cases.</li>
    <li>Mild, moderately severe, and severe cases of depression represented 27%, 25%, and 10% of the total cases respectively.</li>
</ul>

```python
r = PHQ_data['PHQ_range'].value_counts()  
plt.figure(figsize=(12, 9))
ax = sns.barplot(r)
for i in ax.containers:
    ax.bar_label(i)
plt.title('GRAPH SHOWING PHQ DIAGNOSIS COUNT')
plt.xlabel('PHQ SCALE')
plt.ylabel('COUNT')
plt.show()
```

<p>We draw the following conclusions from the PHQ graph:</p>
<ul>
    <li>Most participants were either experiencing mild or moderate depression symptoms.</li>
    <li>68 participants were showing severe depression symptoms while 16 other participants were showing more severe depression symptoms.</li>
</ul>

```python
PHQ_data = pd.concat([PHQ_data, shamiri_data.iloc[:, 27:32]], axis=1)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(20, 15))
fig.suptitle('DEPRESSION SYMPTOMS', c='r')

plt.subplot(2, 2, 1)
c = PHQ_data[['PHQ_range', 'Tribe']].groupby(PHQ_data['Tribe'])
c.get_group('Minority').value_counts(normalize=True, sort=False).plot(label='Minority')
c.get_group('Majority').value_counts(normalize=True, sort=False).plot(label='Majority')
plt.title('PHQ SCALE DISTRIBUTION AMONG THE DIFFERENT TRIBES')
d = np.arange(0, 5)
plt.xticks(ticks=d, labels=('mild', 'moderate', 'moderately-severe', 'severe', 'more-severe'))
plt.xlabel('PHQ SCALE')
plt.ylabel('COUNT AS A % OF TOTAL')
plt.legend()

plt.subplot(2, 2, 2)
g = PHQ_data[['PHQ_range', 'Gender']].groupby(PHQ_data['Gender'])
g.get_group('F').value_counts(normalize=True, sort=False).plot(label='Female')
g.get_group('M').value_counts(normalize=True, sort=False).plot(label='Male')
plt.title('PHQ SCALE DISTRIBUTION AMONG THE DIFFERENT GENDERS')
d = np.arange(0, 5)
plt.xticks(ticks=d, labels=('mild', 'moderate', 'moderately-severe', 'severe', 'more-severe'))
plt.xlabel('PHQ SCALE')
plt.ylabel('COUNT AS A % OF TOTAL')
plt.legend()

plt.subplot(2, 2, 3)
s = PHQ_data[['PHQ_range', 'School']].groupby(PHQ_data['School'])
s.get_group('Olympic').value_counts(normalize=True, sort=False).plot(label='Olympic')
s.get_group('AGHS').value_counts(normalize=True, sort=False).plot(label='AGHS')
s.get_group('AHS').value_counts(normalize=True, sort=False).plot(label='AHS')
s.get_group('Elite').value_counts(normalize=True, sort=False).plot(label='Elite')
s.get_group('Starays').value_counts(normalize=True, sort=False).plot(label='Starays')
plt.title('PHQ SCALE DISTRIBUTION AMONG THE DIFFERENT SCHOOLS')
d = np.arange(0, 5)
plt.xticks(ticks=d, labels=('mild', 'moderate', 'moderately-severe', 'severe', 'more-severe'))
plt.xlabel('PHQ SCALE')
plt.ylabel('COUNT AS A % OF TOTAL')
plt.legend()

plt.subplot(2, 2, 4)
sr = PHQ_data[['PHQ_range', 'School_Resources']].groupby(PHQ_data['School_Resources'])
sr.get_group('Rich').value_counts(normalize=True, sort=False).plot(label='Rich')
sr.get_group('Medium').value_counts(normalize=True, sort=False).plot(label='Medium')
sr.get_group('Poor').value_counts(normalize=True, sort=False).plot(label='Poor')
plt.title('PHQ SCALE DISTRIBUTION AMONG THE DIFFERENT LEVELS OF SCHOOL RESOURCES')
d = np.arange(0, 5)
plt.xticks(ticks=d, labels=('mild', 'moderate', 'moderately-severe', 'severe', 'more-severe'))
plt.xlabel('PHQ SCALE')
plt.ylabel('COUNT AS A % OF TOTAL')
plt.legend()

plt.show()
```

<p>The count data has been standardized by dividing by the respective individual totals to cater for the difference in the total number of participants in each category.</p>
<p>From these graphs we observe that:</p>
<ul>
    <li>Generally, the minority tribe was proportionately more depressed than the majority tribe. However, mild and more extreme cases of depression were more prevalent in the majority tribe.</li>
    <li>The male participants were proportionately exhibiting more mild, moderate, and moderately severe symptoms of depression than the female participants, who were exhibiting more, proportionately, severe, and more severe symptoms of depression.</li>
    <li>Participants from AHS school had proportionately more mild symptoms of depression than the rest of the schools, while those from AGHS school had proportionately more severe symptoms of depression than the rest of the schools.</li>
    <li>Participants from rich school backgrounds had proportionately more mild cases of depression than those from poor and medium backgrounds. However, they also had proportionately more cases of severe depression than the other backgrounds. Also, participants from medium and poor school backgrounds had nearly similar proportions of depression cases.</li>
</ul>

<h1>Analysis of Generalized Anxiety Disorder(GAD)</h1>
<p>To analyze GAD data we will need to create a new and exclusive DataFrame of only GAD data. We will also need to create a new column for the individual scores.</p>

```python
GAD_data = shamiri_data.iloc[ :, 8:15] #We slice the original dataframe to remain only with GAD data
score = GAD_data.sum(axis=1)  #We then find the individual scores of each participant
GAD_data['GAD_score'] = score  #We then create a new column showing the score of each participant
GAD_data
```

```python
print(GAD_data['GAD_score'].min(), GAD_data['GAD_score'].max())    #We confirm whether the scores in the score column are within range
```

<p>We then create bins to classify the individual scores into their respective anxiety disorder diagnosis. We know that the cutoffs for the GAD-7 are 5, 10, and 15, indicating mild, moderate, and severe anxiety, respectively. We also know that GAD scale runs from 0 to 21.</p>

```python
bins = np.arange(0, 21, 5)
scale = ['mild', 'moderate', 'severe', 'more-severe']
GAD_data['GAD_range'] = pd.cut(GAD_data['GAD_score'], bins=bins, labels=scale, include_lowest=True)
GAD_data['GAD_range'].fillna('more-severe', inplace=True)
GAD_data
```

```python
GAD_data['GAD_range'].value_counts(normalize=True, sort=False)   #Proportion of the total
```

<p>The above output shows that:</p> 
<ul>
    <li>Mild anxiety cases was leading with 35% of the total cases while the more severe cases of anxiety was trailing with 9% of the total cases.</li>
    <li>Moderate and severe cases of anxiety represented 33% and 22% of the total cases respectively.</li>
</ul>

```python
h = GAD_data['GAD_range'].value_counts()  #color=('red', 'blue', 'orange', 'yellow', 'black')
plt.figure(figsize=(12, 9))
ax = sns.barplot(h)
for i in ax.containers:
    ax.bar_label(i)
plt.title('GAD Count')
plt.xlabel('GAD Scale')
plt.ylabel('Count')
plt.show()
```

<p>We draw the following conclusions from this GAD graph:</p>
<ul>
    <li>Most participants were either experiencing mild or moderate anxiety symptoms.</li>
    <li>145 participants were showing severe anxiety symptoms while 61 other participants were showing more severe anxiety symptoms.</li>
</ul>
